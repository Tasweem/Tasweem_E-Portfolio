```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("C:\\Users\\tasweemb\\Documents\\Machine Learning Module\\Datasets\\Auto\\auto-mpg.csv")


# Selecting 'weight' as x-axis and 'mpg' as y-axis
x = df['weight'].values
y = df['mpg'].values

# Fitting a polynomial model of degree 3
mymodel = np.poly1d(np.polyfit(x, y, 3))

# Specifying how the line will display
myline = np.linspace(min(x), max(x), 100)

plt.scatter(x, y, color='blue', s=10, label='Data Points')
plt.plot(myline, mymodel(myline), color='red', label='Polynomial Fit')
plt.title('Polynomial Regression of mpg vs weight')
plt.xlabel('Weight')
plt.ylabel('mpg')
plt.legend()
plt.grid(True)
plt.show()
```

Here's the plot showing the relationship between weight and mpg with a third-degree polynomial regression fit.

The blue dots represent the actual data points from the auto-mpg.csv dataset, while the red curve is the polynomial fit. As the weight of the car increases, the miles per gallon (mpg) tends to decrease, which is consistent with our expectations. The polynomial fit captures this trend and provides a more flexible representation compared to a straight-line fit.

```python
from sklearn.metrics import r2_score

print(r2_score(y, mymodel(x)))
```

The R 2 score for this third-degree polynomial regression model is approximately 0.7148. This means that the model explains about 71.48% of the variance in the mpg values.

```python
mpg = mymodel(1700)
print(mpg)
```

For a car weight of 1700 pounds, the predicted miles per gallon (mpg) using our polynomial regression model is approximately 35.79 mpg.
