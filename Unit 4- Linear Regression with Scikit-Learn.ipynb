# Importing the required packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import seaborn as sns
```

```python
# Reading the data

gdp_data=pd.read_csv("C:\\Users\\tasweemb\\Documents\Machine Learning Module\\Datasets\\Global\\Unit04 Global_GDP.csv")
population_data=pd.read_csv("C:\\Users\\tasweemb\\Documents\\Machine Learning Module\\Datasets\\Global\\Unit04 Global_Population.csv")
```

```python
# Take a look at the dataset
gdp_data.head(10)
```

```python
population_data.head(10)
```

```python
# Adjust the years range to 2001-2020 for the computations
adjusted_years = [str(year) for year in range(2001, 2021)]
```

```python
# Convert relevant columns to numeric data type for interpolation
for year in adjusted_years:
    gdp_data[year] = pd.to_numeric(gdp_data[year], errors='coerce')
    population_data[year] = pd.to_numeric(population_data[year], errors='coerce')
    
# Apply linear interpolation to the relevant columns (2001-2020) for both datasets
gdp_data[adjusted_years] = gdp_data[adjusted_years].interpolate(method='linear', axis=1, limit_direction='both')
population_data[adjusted_years] = population_data[adjusted_years].interpolate(method='linear', axis=1, limit_direction='both')
```

```python
# Compute the mean GDP and population for each country from 2001 to 2020
mean_gdp = gdp_data[adjusted_years].mean(axis=1)
mean_population = population_data[adjusted_years].mean(axis=1)

# Add these computed means to the datasets
gdp_data['Mean_GDP_2001_2020'] = mean_gdp
population_data['Mean_Population_2001_2020'] = mean_population

# Display the first few rows of each dataset with the new columns
gdp_data[['Country Name', 'Mean_GDP_2001_2020']].head(), population_data[['Country Name', 'Mean_Population_2001_2020']].head()
```

```python
# Merge the datasets on the "Country Name" column
merged_data = pd.merge(gdp_data[['Country Name', 'Mean_GDP_2001_2020']],
                       population_data[['Country Name', 'Mean_Population_2001_2020']],
                       on='Country Name')

# Compute the mean per capita GDP for each country from 2001 to 2020
merged_data['Mean_Per_Capita_GDP_2001_2020'] = merged_data['Mean_GDP_2001_2020'] / merged_data['Mean_Population_2001_2020']

# Display the merged dataset with the new column
merged_data[['Country Name', 'Mean_GDP_2001_2020', 'Mean_Population_2001_2020', 'Mean_Per_Capita_GDP_2001_2020']].head()
```

```python
# Check for duplicated "Country Name" entries in both datasets
gdp_duplicated = gdp_data[gdp_data['Country Name'].duplicated(keep=False)]
population_duplicated = population_data[population_data['Country Name'].duplicated(keep=False)]

gdp_duplicated[['Country Name']], population_duplicated[['Country Name']]
```

```python
# Remove duplicate entries for Aruba from the population_data dataset
population_data_cleaned = population_data.drop_duplicates(subset='Country Name', keep='first')

# Drop rows with NaN as the "Country Name"
population_data_cleaned = population_data_cleaned.dropna(subset=['Country Name'])

# Merge the datasets again on the "Country Name" column
merged_data_cleaned = pd.merge(gdp_data[['Country Name', 'Mean_GDP_2001_2020']],
                               population_data_cleaned[['Country Name', 'Mean_Population_2001_2020']],
                               on='Country Name')

# Compute the mean per capita GDP for each country from 2001 to 2020
merged_data_cleaned['Mean_Per_Capita_GDP_2001_2020'] = merged_data_cleaned['Mean_GDP_2001_2020'] / merged_data_cleaned['Mean_Population_2001_2020']

# Display the cleaned merged dataset with the new column
merged_data_cleaned[['Country Name', 'Mean_GDP_2001_2020', 'Mean_Population_2001_2020', 'Mean_Per_Capita_GDP_2001_2020']].head()
```

```python
merged_data_cleaned
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Mean Population vs Mean Per Capita GDP
plt.figure(figsize=(12, 7))
plt.scatter(merged_data_cleaned['Mean_Population_2001_2020'], 
            merged_data_cleaned['Mean_Per_Capita_GDP_2001_2020'], 
            color='blue', 
            alpha=0.6)

plt.title('Mean Population vs. Mean Per Capita GDP (2001-2020)')
plt.xlabel('Mean Population (2001-2020)')
plt.ylabel('Mean Per Capita GDP (2001-2020)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Compute the Pearson Correlation Coefficient between Mean Population and Mean Per Capita GDP
correlation_coefficient = merged_data_cleaned['Mean_Population_2001_2020'].corr(merged_data_cleaned['Mean_Per_Capita_GDP_2001_2020'])

correlation_coefficient
```

# LINEAR REGRESSION


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Drop rows with NaN values
merged_data_cleaned.dropna(inplace=True)

# Prepare the data
X = merged_data_cleaned['Mean_Population_2001_2020'].values.reshape(-1, 1)
y = merged_data_cleaned['Mean_GDP_2001_2020'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Extract model parameters
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, y)

slope, intercept, r_squared
```
