```python
import numpy as np
import zipfile
import os
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('/content/hospitality_dataset.xlsx')

# Convert date columns to datetime format
date_columns = ['Created On', 'Arrival Date', 'Departure Date', 'DOB']
for col in date_columns:
    df[col] = pd.to_datetime(df[col])

# Calculate total spending
df['Total Spending'] = df['Total Pkg Rev Excl VAT'] + df['Total Extras']

# Determine high spenders (top 15%)
spending_threshold = df['Total Spending'].quantile(0.85)
df['High Spender'] = (df['Total Spending'] >= spending_threshold).astype(int)

# Calculate the total number of children
df['Total Children'] = df['Total Child 0-3'] + df['Total Child 4-6'] + df['Total Child 7-12'] + df['Total Child 13-17']


# Drop unnecessary columns
columns_to_drop = ['ReservationStayID', 'Property', 'Market Segment Code', 'MTH',
                   'Total Pkg Rev Excl VAT', 'Total Extras', 'Created On',
                   'Arrival Date', 'Departure Date', 'DOB', 'Star Rating']
df.drop(columns=columns_to_drop, inplace=True)

def preprocess_category(category_filter):
    df_filtered = df[df['Grouped Property'] == category_filter]

    # Ensure only significant countries are included
    country_counts = df_filtered['Country'].value_counts()
    total_bookings = len(df_filtered)
    country_percentages = (country_counts / total_bookings) * 100
    significant_countries = country_percentages[country_percentages >= 0.1].index
    df_filtered = df_filtered[df_filtered['Country'].isin(significant_countries)]

    # Encode categorical variables
    df_encoded = pd.get_dummies(df_filtered, columns=['Category', 'Meal Plan Code', 'Room Type',
                                                     'Reason For Stay', 'Source of Business',
                                                     'Booking type'], drop_first=True)

    # Impute missing numeric data
    numeric_columns = df_encoded.select_dtypes(include=[np.number]).columns
    imputer = SimpleImputer(strategy='mean')
    df_imputed_numeric = pd.DataFrame(imputer.fit_transform(df_encoded[numeric_columns]), columns=numeric_columns)
    df_imputed = pd.concat([df_imputed_numeric, df_encoded.drop(columns=numeric_columns).reset_index(drop=True)], axis=1)

    return df_imputed

# Applying preprocessing to specific categories
data_4_star_family = preprocess_category('4-star Family')
data_4_star_adults_only = preprocess_category('4-star Adults Only')
data_3_star_adults_only = preprocess_category('3-star Adults Only')
data_3_star_family = preprocess_category('3-star Family')
data_5_star_adults_only = preprocess_category('5-star Adults Only')

# Print the columns after preprocessing for different categories
print("Columns in df_imputed after preprocessing for 4-star Family:")
print(data_4_star_family.columns)
print("Columns in df_imputed after preprocessing for 4-star Adults Only:")
print(data_4_star_adults_only.columns)
print("Columns in df_imputed after preprocessing for 3-star Family:")
print(data_3_star_family.columns)
print("Columns in df_imputed after preprocessing for 3-star Adults Only:")
print(data_3_star_adults_only.columns)
print("Columns in df_imputed after preprocessing for 5-star Adults Only:")
print(data_5_star_adults_only.columns)
```

# Modelling

MLP Model 4 star family

```python
df_imputed = pd.read_csv('data_4_star_family.csv')

# Prepare features and target
X_4_star_family = df_imputed.drop('High Spender', axis=1)
y_4_star_family = df_imputed['High Spender'].astype(int)

# Identify numeric and non-numeric features
numeric_features = X_4_star_family.select_dtypes(include=[np.number]).columns
non_numeric_features = X_4_star_family.select_dtypes(exclude=[np.number]).columns

# Setup preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), non_numeric_features)
    ])

# Apply preprocessing
X_4_star_family_processed = preprocessor.fit_transform(X_4_star_family)

# Split data
X_train_4_star_family, X_test_4_star_family, y_train_4_star_family, y_test_4_star_family = train_test_split(
    X_4_star_family_processed, y_4_star_family, test_size=0.3, random_state=42, stratify=y_4_star_family)

model_4_star_family = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_4_star_family.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_4_star_family.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_4_star_family = model_4_star_family.fit(
    X_train_4_star_family, y_train_4_star_family,
    epochs=50,
    batch_size=256,
    validation_data=(X_test_4_star_family, y_test_4_star_family),
    verbose=1
)
test_loss_4_star_family, test_accuracy_4_star_family = model_4_star_family.evaluate(X_test_4_star_family, y_test_4_star_family, verbose=2)
print(f"Test Accuracy: {test_accuracy_4_star_family:.4f}")

y_pred_proba_4_star_family = model_4_star_family.predict(X_test_4_star_family)
y_pred_4_star_family = (y_pred_proba_4_star_family >= 0.5).astype(int)

print("Classification Report:")
print(classification_report(y_test_4_star_family, y_pred_4_star_family))
print("Confusion Matrix:")
print(confusion_matrix(y_test_4_star_family, y_pred_4_star_family))
roc_auc_4_star_family = roc_auc_score(y_test_4_star_family, y_pred_proba_4_star_family)
print(f"ROC-AUC Score: {roc_auc_4_star_family:.4f}")
```

k-fold cross validation For MLP Model 4 star family

```python
# Suppress warnings
warnings.filterwarnings("ignore", category=UserWarning)

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracy_scores = []
roc_auc_scores = []
all_classification_reports = []
all_confusion_matrices = []

for train_index, test_index in kfold.split(X_4_star_family, y_4_star_family):

    X_train, X_test = X_4_star_family.iloc[train_index], X_4_star_family.iloc[test_index]
    y_train, y_test = y_4_star_family.iloc[train_index], y_4_star_family.iloc[test_index]

    X_train = X_train.apply(pd.to_numeric, errors='coerce')
    X_test = X_test.apply(pd.to_numeric, errors='coerce')

    X_train = X_train.fillna(0)
    X_test = X_test.fillna(0)

    numeric_features = X_train.select_dtypes(include=[np.number]).columns

    assert not np.isnan(X_train[numeric_features].values).any(), "NaN values found in X_train after processing."
    assert not np.isnan(X_test[numeric_features].values).any(), "NaN values found in X_test after processing."

    scaler = StandardScaler()

    X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
    X_test[numeric_features] = scaler.transform(X_test[numeric_features])

    model = Sequential([
        Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    model.fit(X_train, y_train, epochs=50, batch_size=256, verbose=0)

    y_pred_proba = model.predict(X_test)
    y_pred = (y_pred_proba >= 0.5).astype(int)

    # Calculate metrics
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    if not np.isnan(y_pred_proba).any():
        roc_auc_scores.append(roc_auc_score(y_test, y_pred_proba))
    else:
        print("NaN detected in y_pred_proba, skipping ROC-AUC calculation for this fold.")
        roc_auc_scores.append(np.nan)

    all_classification_reports.append(classification_report(y_test, y_pred))
    all_confusion_matrices.append(confusion_matrix(y_test, y_pred))

print(f"Cross-Validation Accuracy: {np.mean(accuracy_scores):.4f} ± {np.std(accuracy_scores):.4f}")
print(f"Cross-Validation ROC-AUC: {np.nanmean(roc_auc_scores):.4f} ± {np.nanstd(roc_auc_scores):.4f}")

for i, report in enumerate(all_classification_reports):
    print(f"\nFold {i+1} - Classification Report:\n{report}")

for i, matrix in enumerate(all_confusion_matrices):
    print(f"Fold {i+1} - Confusion Matrix:\n{matrix}")
```

```python
# Function to calculate Accuracy Equality
def accuracy_equality(df, protected_col, label_col, prediction_col, threshold):
    privileged_group = df[df[protected_col] == 1]
    unprivileged_group = df[df[protected_col] == 0]

    # Accuracy for privileged and unprivileged groups
    privileged_accuracy = ((privileged_group[prediction_col] >= threshold) == privileged_group[label_col]).mean()
    unprivileged_accuracy = ((unprivileged_group[prediction_col] >= threshold) == unprivileged_group[label_col]).mean()

    accuracy_equality_diff = unprivileged_accuracy - privileged_accuracy
    return accuracy_equality_diff

threshold = 0.5
accuracy_diff_3_star_family = accuracy_equality(df_3_star_family, 'High Spender', 'High Spender', 'Predicted Probabilities_3_Star_Family', threshold)
print(f"3-Star Family - Accuracy Equality Difference: {accuracy_diff_3_star_family:.4f}")

accuracy_diff_3_star_adults = accuracy_equality(df_3_star_adults_only, 'High Spender', 'High Spender', 'Predicted Probabilities_3_Star_Adults_Only', threshold)
accuracy_diff_4_star_adults = accuracy_equality(df_4_star_adults_only, 'High Spender', 'High Spender', 'Predicted Probabilities_4_Star_Adults_Only', threshold)
accuracy_diff_4_star_family = accuracy_equality(df_4_star_family, 'High Spender', 'High Spender', 'Predicted Probabilities_4_Star_Family', threshold)
accuracy_diff_5_star_adults = accuracy_equality(df_5_star_adults_only, 'High Spender', 'High Spender', 'Predicted Probabilities_5_Star_Adults_Only', threshold)

print(f"3-Star Adults Only - Accuracy Equality Difference: {accuracy_diff_3_star_adults:.4f}")
print(f"4-Star Adults Only - Accuracy Equality Difference: {accuracy_diff_4_star_adults:.4f}")
print(f"4-Star Family - Accuracy Equality Difference: {accuracy_diff_4_star_family:.4f}")
print(f"5-Star Adults Only - Accuracy Equality Difference: {accuracy_diff_5_star_adults:.4f}")
```


Fine tune k means with silhouette_score

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Function to fine-tune K-Means based on Silhouette Score
def fine_tune_kmeans(X, max_k=10):
    best_k = 0
    best_score = -1
    silhouette_scores = []

    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
        cluster_labels = kmeans.fit_predict(X)

        # Calculate silhouette score
        score = silhouette_score(X, cluster_labels)
        silhouette_scores.append(score)

        print(f"Silhouette Score for k={k}: {score:.4f}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\nBest k based on Silhouette Score: {best_k} with a score of {best_score:.4f}")

    return best_k, silhouette_scores

# Extract PCA components from data
X_pca = high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']].values

# Fine-tune K-Means and get the best number of clusters
best_k, silhouette_scores = fine_tune_kmeans(X_pca, max_k=10)

# Apply K-Means clustering using the best_k value
kmeans = KMeans(n_clusters=best_k, random_state=42)
high_spenders_pca_df['Cluster'] = kmeans.fit_predict(high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']])

# Plot the silhouette scores for different values of k
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different k (3-Star Family)', fontsize=15)
plt.xlabel('Number of clusters', fontsize=15)
plt.ylabel('Silhouette Score', fontsize=15)
plt.show()

# Print the data with the fine-tuned K-Means clusters
print("Data with fine-tuned K-Means clusters:")
print(high_spenders_pca_df.head())

# Plot the fine-tuned K-Means clustering
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=high_spenders_pca_df,
    x='PCA Component 1',
    y='PCA Component 2',
    hue='Cluster',
    palette='viridis',
    s=100
)
plt.title(f'Fine-Tuned K-Means Clustering (Best k = {best_k}) for High Spenders (3-Star Family)', fontsize=15)
plt.xlabel('PCA Component 1', fontsize=15)
plt.ylabel('PCA Component 2', fontsize=15)
plt.legend(title='Cluster', fontsize=15, title_fontsize=15, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate silhouette samples for the best_k
silhouette_vals = silhouette_samples(X_pca, high_spenders_pca_df['Cluster'])

# Plot silhouette scores for each sample
plt.figure(figsize=(10, 7))
y_lower, y_upper = 0, 0
for i in range(best_k):
    cluster_silhouette_vals = silhouette_vals[high_spenders_pca_df['Cluster'] == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * len(cluster_silhouette_vals), str(i))
    y_lower = y_upper

plt.axvline(x=np.mean(silhouette_vals), color="red", linestyle="--")
plt.title(f'Silhouette Plot for Fine-Tuned K-Means Clustering (k = {best_k})', fontsize=15)
plt.xlabel('Silhouette Coefficient', fontsize=15)
plt.ylabel('Cluster', fontsize=15)
plt.show()
```

```python
df_imputed_full_3_star_family = pd.read_csv('/content/3_star_family_with_clusters.csv')

# Separate the features, excluding the target and cluster labels
X_full_3_star_family = df_imputed_full_3_star_family.drop(['High Spender', 'Cluster'], axis=1, errors='ignore')

# Identify numeric and categorical features
numeric_features = X_full_3_star_family.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_full_3_star_family.select_dtypes(include=['object', 'bool']).columns.tolist()

# Setup the ColumnTransformer for clustering
preprocessor_3_star_family = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

# Fit and transform the features dataset
preprocessor_3_star_family.fit(X_full_3_star_family)
X_full_processed_3_star_family = preprocessor_3_star_family.transform(X_full_3_star_family)

# Add placeholder features if needed
missing_feature_count = 113 - X_full_processed_3_star_family.shape[1]
if missing_feature_count > 0:
    placeholder_features = np.zeros((X_full_processed_3_star_family.shape[0], missing_feature_count))
    X_full_processed_3_star_family = np.hstack((X_full_processed_3_star_family, placeholder_features))

# Check the updated shape to ensure it matches the model's expected input shape
print(f"Updated shape of processed data: {X_full_processed_3_star_family.shape}")

# Get feature names
feature_names = preprocessor_3_star_family.get_feature_names_out()

# Print diagnostics
print(f"Number of feature names: {len(feature_names)}")
print(f"Shape of processed data: {X_full_processed_3_star_family.shape}")

# Ensure that feature names match the number of features
if len(feature_names) != X_full_processed_3_star_family.shape[1]:
    # Adjust the feature names if placeholder features were added
    feature_names = list(feature_names) + [f'placeholder_{i}' for i in range(missing_feature_count)]
    #print(f"Adjusted feature names: {feature_names[-10:]}")

# Define a wrapper function to match the output format expected by LIME
def predict_proba_wrapper(data):
    proba = model_3_star_family.predict(data)
    return np.hstack([1 - proba, proba])

# Generate LIME explanations for Cluster 0
cluster_0_high_spenders_3_star_family = df_imputed_full_3_star_family[(df_imputed_full_3_star_family['Cluster'] == 0) & (df_imputed_full_3_star_family['High Spender'] == 1)]
cluster_0_indices_full_3_star_family = cluster_0_high_spenders_3_star_family.index
X_cluster_0_full_3_star_family = X_full_processed_3_star_family[cluster_0_indices_full_3_star_family, :113]  # Adjusted for the correct input shape

lime_explanations_dir_full_3_star_family_0 = 'lime_explanations_3_star_family_full_cluster_0'
os.makedirs(lime_explanations_dir_full_3_star_family_0, exist_ok=True)

# Initialize the LIME explainer
explainer_3_star_family_full = LimeTabularExplainer(
    training_data=X_full_processed_3_star_family[:, :113],
    feature_names=feature_names[:113],  # Use the correct number of feature names
    class_names=['Low Spender', 'High Spender'],
    mode='classification'
)

# Aggregate feature importance
feature_importance_full_3_star_family_0 = defaultdict(list)
for idx in range(X_cluster_0_full_3_star_family.shape[0]):
    explanation = explainer_3_star_family_full.explain_instance(
        data_row=X_cluster_0_full_3_star_family[idx],
        predict_fn=predict_proba_wrapper,
        num_features=200
    )
    explanation.save_to_file(f'{lime_explanations_dir_full_3_star_family_0}/lime_explanation_{cluster_0_indices_full_3_star_family[idx]}.html')
    for feature, importance in explanation.as_list(label=1):
        feature_importance_full_3_star_family_0[feature].append(importance)

avg_feature_importance_full_3_star_family_0 = {feature: np.mean(importance_list) for feature, importance_list in feature_importance_full_3_star_family_0.items()}
importance_df_full_3_star_family_0 = pd.DataFrame(list(avg_feature_importance_full_3_star_family_0.items()), columns=['Feature', 'Average Importance'])
importance_df_full_3_star_family_0.to_csv('aggregated_feature_importance_3_star_family_full_cluster_0.csv', index=False)

zip_filename_3_star_family_0 = 'lime_explanations_3_star_family_full_cluster_0.zip'
with zipfile.ZipFile(zip_filename_3_star_family_0, 'w') as zipf:
    for file in os.listdir(lime_explanations_dir_full_3_star_family_0):
        zipf.write(os.path.join(lime_explanations_dir_full_3_star_family_0, file))
    zipf.write('aggregated_feature_importance_3_star_family_full_cluster_0.csv')

print(f"Total LIME explanations saved for Cluster 0: {X_cluster_0_full_3_star_family.shape[0]}")
print(f"LIME explanations and aggregated feature importance saved to {zip_filename_3_star_family_0}")
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Function to fine-tune K-Means based on Silhouette Score
def fine_tune_kmeans(X, max_k=10):
    best_k = 0
    best_score = -1
    silhouette_scores = []

    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
        cluster_labels = kmeans.fit_predict(X)

        # Calculate silhouette score
        score = silhouette_score(X, cluster_labels)
        silhouette_scores.append(score)

        print(f"Silhouette Score for k={k}: {score:.4f}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\nBest k based on Silhouette Score: {best_k} with a score of {best_score:.4f}")
    return best_k, silhouette_scores

# Extract the PCA components for clustering
X_pca_3star_adults = high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']].values

# Fine-tune K-Means and get the best number of clusters
best_k_3star_adults, silhouette_scores_3star_adults = fine_tune_kmeans(X_pca_3star_adults, max_k=10)

# Apply K-Means clustering using the best_k value
kmeans = KMeans(n_clusters=best_k_3star_adults, random_state=42)
high_spenders_pca_df['Cluster'] = kmeans.fit_predict(high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']])

# Plot silhouette scores for different values of k
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores_3star_adults, marker='o')
plt.title('Silhouette Scores for Different k (3-Star Adults Only)', fontsize=15)
plt.xlabel('Number of clusters', fontsize=15)
plt.ylabel('Silhouette Score', fontsize=15)
plt.show()

# Plot K-Means clustering with the optimal number of clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=high_spenders_pca_df,
    x='PCA Component 1',
    y='PCA Component 2',
    hue='Cluster',
    palette='viridis',
    s=100
)
plt.title(f'Fine-Tuned K-Means Clustering (Best k = {best_k_3star_adults}) for 3-Star Adults Only', fontsize=15)
plt.xlabel('PCA Component 1', fontsize=15)
plt.ylabel('PCA Component 2', fontsize=15)
plt.legend(title='Cluster', fontsize=15, title_fontsize=15, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate silhouette samples for the best_k
silhouette_vals_3star_adults = silhouette_samples(X_pca_3star_adults, high_spenders_pca_df['Cluster'])

# Plot silhouette scores for each sample
plt.figure(figsize=(10, 7))
y_lower, y_upper = 0, 0
for i in range(best_k_3star_adults):
    cluster_silhouette_vals = silhouette_vals_3star_adults[high_spenders_pca_df['Cluster'] == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * len(cluster_silhouette_vals), str(i))
    y_lower = y_upper

plt.axvline(x=np.mean(silhouette_vals_3star_adults), color="red", linestyle="--")
plt.title(f'Silhouette Plot for Fine-Tuned K-Means Clustering (k = {best_k_3star_adults})', fontsize=15)
plt.xlabel('Silhouette Coefficient', fontsize=15)
plt.ylabel('Cluster', fontsize=15)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Function to fine-tune K-Means based on Silhouette Score
def fine_tune_kmeans(X, max_k=10):
    best_k = 0
    best_score = -1
    silhouette_scores = []

    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
        cluster_labels = kmeans.fit_predict(X)

        # Calculate silhouette score
        score = silhouette_score(X, cluster_labels)
        silhouette_scores.append(score)

        print(f"Silhouette Score for k={k}: {score:.4f}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\nBest k based on Silhouette Score: {best_k} with a score of {best_score:.4f}")
    return best_k, silhouette_scores

# Extract the PCA components for clustering (4-Star Family dataset)
X_pca_4star_family = high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']].values

# Fine-tune K-Means and get the best number of clusters
best_k_4star_family, silhouette_scores_4star_family = fine_tune_kmeans(X_pca_4star_family, max_k=10)

# Apply K-Means clustering using the best_k value
kmeans = KMeans(n_clusters=best_k_4star_family, random_state=42)
high_spenders_pca_df['Cluster'] = kmeans.fit_predict(high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']])

# Plot silhouette scores for different values of k
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores_4star_family, marker='o')
plt.title('Silhouette Scores for Different k (4-Star Family)', fontsize=15)
plt.xlabel('Number of clusters', fontsize=15)
plt.ylabel('Silhouette Score', fontsize=15)
plt.show()

# Plot K-Means clustering with the optimal number of clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=high_spenders_pca_df,
    x='PCA Component 1',
    y='PCA Component 2',
    hue='Cluster',
    palette='viridis',
    s=100
)
plt.title(f'Fine-Tuned K-Means Clustering (Best k = {best_k_4star_family}) for 4-Star Family', fontsize=15)
plt.xlabel('PCA Component 1', fontsize=15)
plt.ylabel('PCA Component 2', fontsize=15)
plt.legend(title='Cluster', fontsize=15, title_fontsize=15, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate silhouette samples for the best_k
silhouette_vals_4star_family = silhouette_samples(X_pca_4star_family, high_spenders_pca_df['Cluster'])

# Plot silhouette scores for each sample
plt.figure(figsize=(10, 7))
y_lower, y_upper = 0, 0
for i in range(best_k_4star_family):
    cluster_silhouette_vals = silhouette_vals_4star_family[high_spenders_pca_df['Cluster'] == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * len(cluster_silhouette_vals), str(i))
    y_lower = y_upper

plt.axvline(x=np.mean(silhouette_vals_4star_family), color="red", linestyle="--")
plt.title(f'Silhouette Plot for Fine-Tuned K-Means Clustering (k = {best_k_4star_family})', fontsize=15)
plt.xlabel('Silhouette Coefficient', fontsize=15)
plt.ylabel('Cluster', fontsize=15)
plt.show()
```

```python
df_imputed_full = pd.read_csv('data_5_star_adults_only_with_predictions.csv')

threshold = 0.5
high_spenders = df_imputed_full[df_imputed_full['Predicted Probabilities_5_Star_Adults_Only'] >= threshold]

numeric_features = high_spenders.select_dtypes(include=['int64', 'float64']).columns.tolist()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(high_spenders[numeric_features])

pca = PCA(n_components=2)
high_spenders_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(data=high_spenders_pca, columns=['PCA Component 1', 'PCA Component 2'])

high_spenders_pca_df = pd.concat([high_spenders.reset_index(drop=True), pca_df], axis=1)

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']])
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal k', fontsize=15)
plt.xlabel('Number of clusters', fontsize=15)
plt.ylabel('WCSS', fontsize=15)
plt.show()

optimal_clusters = 3

kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
high_spenders_pca_df['Cluster'] = kmeans.fit_predict(high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']])

high_spenders_pca_df.to_csv('5_star_adults_only_with_clusters.csv', index=False)

cluster_analysis = high_spenders_pca_df.groupby('Cluster').agg({
    'High Spender': ['mean', 'count'],
    'Predicted Probabilities_5_Star_Adults_Only': 'mean'
}).reset_index()

cluster_analysis.columns = ['Cluster', 'High Spender Rate', 'Count in Cluster', 'Average Predicted Probability']
cluster_analysis['High Spender Rate (%)'] = cluster_analysis['High Spender Rate'] * 100

print("Cluster Analysis:")
print(cluster_analysis)

for index, row in cluster_analysis.iterrows():
    print(f"\nCluster {int(row['Cluster'])}:")
    print(f"High Spender Rate: {row['High Spender Rate (%)']:.2f}%")
    print(f"Count in Cluster: {int(row['Count in Cluster'])}")
    print(f"Average Predicted Probability: {row['Average Predicted Probability']:.4f}")
```

Fine tuned k means with silhouette_score

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


# Function to fine-tune K-Means based on Silhouette Score
def fine_tune_kmeans(X, max_k=10):
    best_k = 0
    best_score = -1
    silhouette_scores = []

    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
        cluster_labels = kmeans.fit_predict(X)

        # Calculate silhouette score
        score = silhouette_score(X, cluster_labels)
        silhouette_scores.append(score)

        print(f"Silhouette Score for k={k}: {score:.4f}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\nBest k based on Silhouette Score: {best_k} with a score of {best_score:.4f}")
    return best_k, silhouette_scores

# Extract the PCA components for clustering (5-Star Adults Only dataset)
X_pca_5star_adults = high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']].values

# Fine-tune K-Means and get the best number of clusters
best_k_5star_adults, silhouette_scores_5star_adults = fine_tune_kmeans(X_pca_5star_adults, max_k=10)

# Apply K-Means clustering using the best_k value
kmeans = KMeans(n_clusters=best_k_5star_adults, random_state=42)
high_spenders_pca_df['Cluster'] = kmeans.fit_predict(high_spenders_pca_df[['PCA Component 1', 'PCA Component 2']])

# Plot silhouette scores for different values of k
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores_5star_adults, marker='o')
plt.title('Silhouette Scores for Different k (5-Star Adults Only)', fontsize=15)
plt.xlabel('Number of clusters', fontsize=15)
plt.ylabel('Silhouette Score', fontsize=15)
plt.show()

# Plot K-Means clustering with the optimal number of clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=high_spenders_pca_df,
    x='PCA Component 1',
    y='PCA Component 2',
    hue='Cluster',
    palette='viridis',
    s=100
)
plt.title(f'Fine-Tuned K-Means Clustering (Best k = {best_k_5star_adults}) for 5-Star Adults Only', fontsize=15)
plt.xlabel('PCA Component 1', fontsize=15)
plt.ylabel('PCA Component 2', fontsize=15)
plt.legend(title='Cluster', fontsize=15, title_fontsize=15, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate silhouette samples for the best_k
silhouette_vals_5star_adults = silhouette_samples(X_pca_5star_adults, high_spenders_pca_df['Cluster'])

# Plot silhouette scores for each sample
plt.figure(figsize=(10, 7))
y_lower, y_upper = 0, 0
for i in range(best_k_5star_adults):
    cluster_silhouette_vals = silhouette_vals_5star_adults[high_spenders_pca_df['Cluster'] == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * len(cluster_silhouette_vals), str(i))
    y_lower = y_upper

plt.axvline(x=np.mean(silhouette_vals_5star_adults), color="red", linestyle="--")
plt.title(f'Silhouette Plot for Fine-Tuned K-Means Clustering (k = {best_k_5star_adults})', fontsize=15)
plt.xlabel('Silhouette Coefficient', fontsize=15)
plt.ylabel('Cluster', fontsize=15)
plt.show()
```
